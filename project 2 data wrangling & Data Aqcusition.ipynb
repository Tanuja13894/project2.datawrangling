```python
lst = ['instant', 'atemp', 'hum', 'windspeed',
            'casual', 'registered', 'cnt']

# Calling DataFrame constructor on list
df = pd.DataFrame(lst)
print(df)
```

```python
# Removing rows with missing values
df_cleaned = df.dropna()

# Displaying the DataFrame after removing missing values
print("\nDataFrame after removing rows with missing values:")
print(df_cleaned)
```

```python
#  Mean, Median, and Mode Imputation
mean_imputation = dataset_3['casual'].fillna(dataset_3['casual'].mean())
median_imputation = dataset_3['casual'].fillna(dataset_3['casual'].median())
mode_imputation = dataset_3['casual'].fillna(dataset_3['casual'].mode().iloc[0])

print("\nImputation using Mean:")
print(mean_imputation)

print("\nImputation using Median:")
print(median_imputation)

print("\nImputation using Mode:")
print(mode_imputation)
```

```python
# Forward and Backward Fill
forward_fill = dataset_3['weekday'].fillna(method='ffill')
backward_fill = dataset_3['weekday'].fillna(method='bfill')

print("\nForward Fill:")
print(forward_fill)

print("\nBackward Fill:")
print(backward_fill)
```

```python
linear_interpolation = dataset_3['hum'].interpolate(method='linear')
quadratic_interpolation = dataset_3['hum'].interpolate(method='quadratic')

print("\nLinear Interpolation:")
print(linear_interpolation)

print("\nQuadratic Interpolation:")
print(quadratic_interpolation)
```
